```python
import pandas as pd
import sqlite3
from sqlalchemy import create_engine
```

```python
df = pd.read_csv('Resources/airbnb.xl.csv')
df
```

```python
df = df.dropna(how = 'any')
df.info()
```

```python
conn = sqlite3.connect('Airbnb.sqlite')
df.to_sql(name='Airbnb', con=conn, if_exists='replace', index=False)

```

```python
engine = create_engine("sqlite:///Airbnb.sqlite")
# reflect an existing database into a new model
```

```python
Base = automap_base()
# reflect the tables
Base.prepare(autoload_with=engine)
Base.classes.keys()
pd.read_sql("Select * from Airbnb", con = conn)
# conn.close()
```
